```python
#and finally removing the indescribadle characters
A = [')', ',', ';']
B = ['']
players['player_id'] = players['player_id'].str.replace('(','')
for i in range(0, 3):
    players['cadastrado_em']= players['cadastrado_em'].str.replace(A[i], B[0])
players
```

```python
def pre_processing(data_base, paramA, paramB):
    """read_table"""
    params = pd.read_table(data_base)
    params = params.drop([0,1])
    params = pd.DataFrame(
        params[params.columns[0]].
        str.
        split(',',1).
        tolist(), 
        columns=[paramA, paramB]
    )
    A = [')', ',', ';']
    B = ['']
    params[paramA] = params[paramA].str.replace('(', '')
    for i in range(0,3):
        params[paramB] = params[paramB].str.replace(A[i], B[0])
    return(params)
```

```python
pre_processing('players.sql', 'player_id', 'cadastrado_em')
```
