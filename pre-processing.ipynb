# In this notebook I am going to realize the dataset pre-processing.

```python
import pandas as pd
players = pd.read_table('players.sql')
matchmaking_stats = pd.read_table('matchmaking_stats.sql')
monthly_stats = pd.read_table('monthly_stats.sql')
```

```python
#creating the pre_processing function
def pre_processing(data_base, lista_colunas, drop=0):
    params = pd.read_table(data_base)
    if drop == 1:
        params = params.drop([0,1])
    params = pd.DataFrame(
        params[params.columns[0]].
        str.
        split(',',len(lista_colunas)-1).
        tolist(), 
        columns=lista_colunas
    )
    A = [')', ',', ';']
    B = ['']
    params[lista_colunas[0]] = params[lista_colunas[0]].str.replace('(', '')
    params[lista_colunas[-1]] = params[lista_colunas[-1]].str.replace(')', '')
    params[lista_colunas[-1]] = params[lista_colunas[-1]].str.replace(',', '')
    params[lista_colunas[-1]] = params[lista_colunas[-1]].str.replace(';', '')
    for i in range(0,3):
        params[lista_colunas[1]] = params[lista_colunas[1]].str.replace(A[i], B[0])
    return(params)
```

```python
lista_matchmaking = ['player_id', 'matches_played', 'total_wins', 
             'total_loss', 'mm_points', 'kills', 
             'death', 'assist', 'hs', 
             'created_at', 'updated_at']

matchmaking_stats = pre_processing('matchmaking_stats.sql', lista_matchmaking, 1)
matchmaking_stats
```

# monthly_stats

```python
lista_monthly_stats = ['player_id', 'game_mode', 'ref_date', 
         'total_played', 'kills', 'deaths',
         'assists', 'wins', 'loss', 
         'headshots']

monthly_stats = pre_processing('monthly_stats.sql', lista_monthly_stats)
monthly_stats
```

# Changing DF types

* # Players

```python
players['player_id'] = pd.to_numeric(players['player_id'])
players['cadastrado_em'] = pd.to_datetime(players['cadastrado_em'])
players
```

* # matchmaking_stats

```python
lista_matchmaking_stats = ['player_id', 'matches_played', 'total_wins', 
                           'total_loss', 'mm_points', 'kills', 
                           'death', 'assist', 'hs']
date_list = ['created_at', 'updated_at']
for i in range(len(lista_matchmaking_stats)):
    matchmaking_stats[lista_matchmaking_stats[i]] = pd.to_numeric(matchmaking_stats[lista_matchmaking_stats[i]])

for i in range(len(date_list)):
    matchmaking_stats[date_list[i]] = pd.to_datetime(matchmaking_stats[date_list[i]])

matchmaking_stats
```

* # monthly_stats

```python
lista_monthly_stats = ['player_id', 'total_played', 
                       'kills', 'deaths', 'assists', 
                       'wins', 'loss', 'headshots']
for i in range(len(lista_monthly_stats)):
    monthly_stats[lista_monthly_stats[i]] = pd.to_numeric(monthly_stats[lista_monthly_stats[i]])    

monthly_stats['ref_date'] = pd.to_datetime(monthly_stats['ref_date'])
monthly_stats
```
