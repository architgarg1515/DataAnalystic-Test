# In this notebook I am going to realize the dataset pre-processing.

```python
import pandas as pd
players = pd.read_table('players.sql')
matchmaking_stats = pd.read_table('matchmaking_stats.sql')
monthly_stats = pd.read_table('monthly_stats.sql')
```

```python
players
```

```python
def pre_processing(data_base, paramA, paramB):
    """read_table"""
    params = pd.read_table(data_base)
    params = params.drop([0,1])
    params = pd.DataFrame(
        params[params.columns[0]].
        str.
        split(',',1).
        tolist(), 
        columns=[paramA, paramB]
    )
    A = [')', ',', ';']
    B = ['']
    params[paramA] = params[paramA].str.replace('(', '')
    for i in range(0,3):
        params[paramB] = params[paramB].str.replace(A[i], B[0])
    return(params)
```

```python
players = pre_processing('players.sql', 'player_id', 'cadastrado_em')
players
```

```python
A = [')', ',', ';']
B = ['']
matchmaking_stats['player_id'] = matchmaking_stats['player_id'].str.replace('(','')
for i in range(0, 3):
    matchmaking_stats['updated_at']= matchmaking_stats['updated_at'].str.replace(A[i], B[0])
matchmaking_stats
```
