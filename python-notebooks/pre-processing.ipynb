# In this notebook I am going to realize the dataset pre-processing.

```python
import pandas as pd
players = pd.read_table('../sql/players.sql')
matchmaking_stats = pd.read_table('../sql/matchmaking_stats.sql')
monthly_stats = pd.read_table('../sql/monthly_stats.sql')
```

```python
#creating the pre_processing function
def pre_processing(data_base, lista_colunas, drop=0):
    params = pd.read_table(data_base)
    if drop == 1:
        params = params.drop([0,1])
    params = pd.DataFrame(
        params[params.columns[0]].
        str.
        split(',',len(lista_colunas)-1).
        tolist(), 
        columns=lista_colunas
    )
    A = [')', ',', ';']
    B = ['']
    params[lista_colunas[0]] = params[lista_colunas[0]].str.replace('(', '')
    params[lista_colunas[-1]] = params[lista_colunas[-1]].str.replace(')', '')
    params[lista_colunas[-1]] = params[lista_colunas[-1]].str.replace(',', '')
    params[lista_colunas[-1]] = params[lista_colunas[-1]].str.replace(';', '')
    for i in range(0,3):
        params[lista_colunas[1]] = params[lista_colunas[1]].str.replace(A[i], B[0])
    return(params)
```

# Inserting these values into a database:

```python
import os
import pymysql
from dotenv import load_dotenv
load_dotenv()

HOST = os.getenv('HOST')
USER = os.getenv('MYSQL_USER')
PASSWORD = os.getenv('MYSQL_ROOT_PASSWORD')
DB = os.getenv('MYSQL_DATABASE')

connection = pymysql.connect(host=HOST, 
                      user = USER,
                     password = PASSWORD,
                     db = DB)

cursor = connection.cursor()
```

```python
import pandas as pd
players
```

```python
def varre_player(arg1, arg2):
    query = ("INSERT INTO players (player_id, cadastrado_em) VALUES (%s, %s)")
    data = (arg1,arg2)
    #print(query)
    return cursor.execute(query, data)
```

```python
n = 0
for index, row in players.iterrows():
    if n == 10:
        connection.commit()
        n = 0
        print('Mais dez inseridos!')
    row['cadastrado_em'] = row['cadastrado_em'].strftime("%Y/%m/%d %H:%M:%S")
    varre_player(row['player_id'], row['cadastrado_em'])
    n = n + 1
connection.commit()
print('jóia')
```

```python
def varre_monthly_stats(arg1, arg2, arg3, arg4, arg5, arg6, arg7, arg8, arg9, arg10):
    query = ("INSERT INTO player_monthly_stats (player_id, game_mode, ref_date, total_played, kills, deaths, assists, wins, loss, headshots) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)")
    data = (arg1,arg2, arg3, arg4, arg5, arg6, arg7, arg8, arg9, arg10)
    #print(query)
    return cursor.execute(query, data)
```

```python
m = 0
for index, row in monthly_stats.iterrows():
    if m == 10:
        connection.commit()
        m = 0
        print('Mais dez registros')
    row['ref_date'] = row['ref_date'].strftime("%Y/%m/%d %H:%M:%S")
    varre_monthly_stats(row['player_id'], row['game_mode'], row['ref_date'],
         row['total_played'], row['kills'], row['deaths'],
         row['assists'], row['wins'], row['loss'],
         row['headshots'])
    m = m + 1
connection.commit()
print('jóia')
```

```python
def varre_matchmaking_stats(arg1, arg2, arg3, arg4, arg5, arg6, arg7, arg8, arg9, arg10, arg11):
    query = ("INSERT INTO matchmaking_stats_summary (player_id, matches_played, total_wins, total_loss, mm_points, kills, death, assist, hs, created_at, updated_at) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)")
                                                    #(player_id, matches_played, total_wins, total_loss, mm_points, kills, death, assist, hs, created_at, updated_at)
    data = (arg1, arg2, arg3, arg4, arg5, arg6, arg7, arg8, arg9, arg10, arg11)
    return cursor.execute(query, data)
```

```python
m = 0
for index, row in matchmaking_stats.iterrows():
    if m == 10:
        connection.commit()
        m = 0
        print('Mais dez registros')
    row['created_at'] = row['created_at'].strftime("%Y/%m/%d %H:%M:%S")
    row['updated_at'] = row['updated_at'].strftime("%Y/%m/%d %H:%M:%S")
    
    varre_matchmaking_stats(row['player_id'], row['matches_played'], row['total_wins'],
                            
         row['total_loss'], row['mm_points'], row['kills'],
                            
         row['death'], row['assist'], row['hs'],
                            
         row['created_at'], row['updated_at'])
    m = m + 1
connection.commit()
print('jóia')
```
